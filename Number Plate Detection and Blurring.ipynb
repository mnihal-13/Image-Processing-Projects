## Russian License Plate Blurring

Welcome to your object detection project! I used Haar Cascades to blur license plates detected in an image!

Russians are famous for having some of the most entertaining DashCam footage on the internet (I encourage you to Google Search "Russian DashCam"). Unfortunately a lot of the footage contains license plates, perhaps we could help out and create a license plat blurring tool.

OpenCV comes with a Russian license plate detector .xml file that we can use.
(unfortunately, it does not come with license detectors for other countries!)

### Import Necessary libraries

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
display(img)
```

### Load the 'haarcascade_russian_plate_number.xml' file.

```python
plate_cascade = cv2.CascadeClassifier('haarcascade_russian_plate_number.xml')
```

Created a function that takes in an image and draws a rectangle around what it detects to be a license plate. I just drawing a rectangle around it for now, later on we will adjust this function to blur.

```python
def detect_plate(img):  
    plate_img = img.copy()

    plate_rects = plate_cascade.detectMultiScale(plate_img, minNeighbors = 4) 

    for (x,y,w,h) in plate_rects: 
        cv2.rectangle(plate_img, (x,y), (x+w,y+h), (0,0,255), 3) 

    return plate_img
```

```python
result = detect_plate(img)
```

```python
display(result)
```

### Blur licence plate

Edit the function so that is effectively blurs the detected plate, instead of just drawing a rectangle around it. Here are the steps we might want to take:**

1. The hardest part is converting the (x,y,w,h) information into the dimension values you need to grab an ROI (somethign we covered in the lecture 01-Blending-and-Pasting-Images. It's simply [Numpy Slicing](https://stackoverflow.com/questions/15589517/how-to-crop-an-image-in-opencv-using-python), we just need to convert the information about the top left corner of the rectangle and width and height, into indexing position values.
2. Once we grabbed the ROI using the (x,y,w,h) values returned, we want to blur that ROI. We can use cv2.medianBlur for this.
3. Now that we have a blurred version of the ROI (the license plate) we want to paste this blurred image back on to the original image at the same original location. Simply using Numpy indexing and slicing to reassign that area of the original image to the blurred roi.

```python
def detect_and_blur_plate(img):
    plate_img = img.copy()
    roi = img.copy()

    plate_rects = plate_cascade.detectMultiScale(plate_img, minNeighbors=4) 
    
    for (x,y,w,h) in plate_rects: 
        
        roi = roi[y:y+h,x:x+w]
        blurred_roi = cv2.medianBlur(roi,7)
        
        plate_img[y:y+h,x:x+w] = blurred_roi
        
    return plate_img
    
```

```python
result = detect_and_blur_plate(img)
```
